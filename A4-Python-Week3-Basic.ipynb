```python
print("BASIC QUESTION 8:")
plt.matshow(df.corr())
plt.show()

corr = df.drop('Survived', axis=1).corr() 
plt.figure(figsize=(12, 10))

sns.heatmap(corr[(corr >= 0.5) | (corr <= -0.4)], 
            cmap='viridis', vmax=1.0, vmin=-1.0, linewidths=0.1,
            annot=True, annot_kws={"size": 8}, square=True);
```

```python
print("BASIC QUESTION 9:")
print("Based on my understanding of the titanic tragedy, wherein preferential treatment in the form of access to evacuation from the sinking ship was provided to the wealthy, and women and children in particular, I expect 'sex' and class of ticket 'pclass' to be of interest for deeper analysis within this data set. ")
print('\n')
print("Survival based on gender:")
print("Female survivors:",'\t',df[(df["Survived"]==1)&(df["Sex"]=='female')].count()["Sex"])
print("Female casualties:",'\t',df[(df["Survived"]==0)&(df["Sex"]=='female')].count()["Sex"])
print("Percentage of females who survived:",'{0:.0%}'.format(df[(df["Survived"]==1)&(df["Sex"]=='female')].count()["Sex"] / df[(df["Sex"]=='female')].count()["Sex"]))
print("Male survivors:",'\t',df[(df["Survived"]==1)&(df["Sex"]=='male')].count()["Sex"])
print("Male casualties:",'\t',df[(df["Survived"]==0)&(df["Sex"]=='male')].count()["Sex"])
print("Percentage of males who survived:",'{0:.0%}'.format(df[(df["Survived"]==1)&(df["Sex"]=='male')].count()["Sex"] / df[(df["Sex"]=='male')].count()["Sex"]))
print('\n')
print("Survival based on ticket class:")
print("Class 1 survivors:",'\t',df[(df["Survived"]==1)&(df["Pclass"]==1)].count()["Pclass"])
print("Class 1 casualties:",'\t',df[(df["Survived"]==0)&(df["Pclass"]==1)].count()["Pclass"])
print("Percentage of Class 1 passengers who survived:",'{0:.0%}'.format(df[(df["Survived"]==1)&(df["Pclass"]==1)].count()["Pclass"] / df[(df["Pclass"]==1)].count()["Pclass"]))
print("Class 2 survivors:",'\t',df[(df["Survived"]==1)&(df["Pclass"]==2)].count()["Pclass"])
print("Class 2 casualties:",'\t',df[(df["Survived"]==0)&(df["Pclass"]==2)].count()["Pclass"])
print("Percentage of Class 2 passengers who survived:",'{0:.0%}'.format(df[(df["Survived"]==1)&(df["Pclass"]==2)].count()["Pclass"] / df[(df["Pclass"]==2)].count()["Pclass"]))
print("Class 3 survivors:",'\t',df[(df["Survived"]==1)&(df["Pclass"]==3)].count()["Pclass"])
print("Class 3 casualties:",'\t',df[(df["Survived"]==0)&(df["Pclass"]==3)].count()["Pclass"])
print("Percentage of Class 3 passengers who survived:",'{0:.0%}'.format(df[(df["Survived"]==1)&(df["Pclass"]==3)].count()["Pclass"] / df[(df["Pclass"]==3)].count()["Pclass"]))

```
