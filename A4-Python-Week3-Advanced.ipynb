```python
print("ADVANCED QUESTION 2:")

df = pd.read_csv('train_titanic.csv')
df.fillna(round(df.mean(),1), inplace = True)
df
```

```python
print("ADVANCED QUESTION 3:")

import matplotlib.pyplot as plt
x=df["Age"]
y=df["Fare"]

plt.scatter(x,y)
plt.title('Scatter plot of Age to Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()

print("Interpretation:")
print("Based on the scatter plot data, it is clear that age was not a determining factor for fares.  The majority of fare values are distributed across all ages.")
```

```python
print("ADVANCED QUESTION 4:")

print("Values for 'Survived' column - Ungrouped:")
print(df['Survived'].describe()[['count','mean','std']])
print('\n')
print("Values for 'Survived' column - Grouped by 'Survived' values:")
print(df.groupby('Survived')['Survived'].describe()[['count','mean','std']])

```

```python
print("ADVANCED QUESTION 5 - Part 1:")

import matplotlib.pyplot as plt
def classCorrelation():
    df_num = df.select_dtypes(include = ['float64', 'int64'])

    corr = df_num.drop('Survived', axis=1).corr()
    plt.figure(figsize=(12, 10))


    sns.heatmap(corr[(corr >= 0.5) | (corr <= -0.4)], 
            cmap='viridis', vmax=1.0, vmin=-1.0, linewidths=0.1,
            annot=True, annot_kws={"size": 8}, square=True)

  
classCorrelation()
print("As indicated in the correlation matrix, the passenger class 'Pclass' attribute is the numeric attribute most correlated with survival.")

```

```python
print("ADVANCED QUESTION 5 - Part 2:")

import matplotlib.pyplot as plt

def fareCorrelation():
    x=df["Pclass"]
    y=df["Fare"]

    plt.scatter(x,y)
    plt.title('Scatter plot of Pclass to Fare')
    plt.xlabel('Pclass')
    plt.ylabel('Fare')
    plt.show()
    
fareCorrelation()
print("As indicated in the scatter plot, the 'Fare' attribute is directly correlated with 'Pclass'.")
print("Therefore, the top 3 attributes correlated with survival are:  'Sex', 'Pclass', 'Fare'")
    
```
